```python
import pandas as pd
```

```python
airport_small = pd.read_csv('monthly_reports_2024/report_2024_airport.csv')
```

```python
airport_all = pd.read_csv('monthly_reports_2024/airports_final.csv')
```

```python
airport_small.head()
```

```python
airport_all.head()
```

```python
airport = pd.merge(airport_small, airport_all,on='AIRPORT_CODE')
```

```python
airport
```

```python
airport['coordinates'] = list(zip(
    airport.LATITUDE.apply(lambda x: round(x,2)),
    airport.LONGITUDE.apply(lambda x: round(x, 2))
    ))
```

```python
airport.info()
```

```python
airport_name = airport[['AIRPORT_CODE', 'AIRPORT_NAME', 'coordinates']]
```
